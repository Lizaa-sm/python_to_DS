* ``**init_notebook_mode(connected=True):**`` loads Plotlyâ€™s JavaScript resources from the internet.
* ``**init_notebook_mode(connected=False):**`` loads the resources offline, without requiring an internet connection (uses local files). After running this, you can display Plotly charts directly inside Jupyter Notebook without needing to call ``fig.show()``.

```python
init_notebook_mode(connected=True)
```

* **from** ``plotly.offline`` **import** ``init_notebook_mode``: Comes from **Plotly** itself and enables Plotly charts to be displayed in Jupyter (either online or offline).
* ``cf.go_offline()`` : Comes from **Cufflinks** and specifically tells Cufflinks to use Plotly in offline mode, so charts can be rendered directly inside Jupyter Notebook without an internet connection.

```python
# Presenting Data
df = pd.DataFrame(np.random.randn(1000,4), columns='A B C D'.split())
df.head()
```

```python
# Creating another Data Frame
df_II = pd.DataFrame({'Columns':['A', 'B', 'C', 'D'], 'Values':[10, 20, 30, 40]})
df_II
```

```python
df.iplot()
# You can do
# df.iplot(kind='scatter', x='A', y='B', mode='markers')
```

```python
# You can do:
df_II.iplot(kind='bar', x='Columns', y='Values', colors='red')
df.sum().iplot(kind='bar', colors='green')
df.iplot(kind='box')
```

```python
df_III = pd.DataFrame({'x':[1, 2, 3, 4, 5],'y':[10, 20, 30, 20, 10], 'z':[5,4,3,2,1] })

df_III.iplot(kind='surface')
```
