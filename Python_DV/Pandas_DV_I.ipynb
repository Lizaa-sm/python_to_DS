* A **histogram** is a chart that displays the distribution of a set of numerical data. It divides the values into intervals (called bins) and uses bars to represent how often the data falls within each interval. This allows us to quickly see how the data is spread, and to identify concentrations, symmetry, skewness, or outliers.

```python
tb_I['B'].hist(bins=50)
# Another way
tb_I['A'].plot(kind='hist', bins=200)
```

```python
tb_II.plot.area(alpha=0.3)
```

```python
ax = tb_II.plot.bar(stacked=True, figsize=(8, 5), alpha=0.8)
ax.set_title('Bar Plot Stacked de tb_II') 
```

```python
tb_I.plot.line(y='A', figsize=(12,3), lw=1)
```

```python
tb_I.plot.scatter(x='A', y='B')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax3d = fig.add_subplot(111, projection='3d')
ax3d.scatter(tb_I['A'], tb_I['B'], tb_I['C'], facecolor='green')
ax3d.set_xlabel('A')
ax3d.set_ylabel('B')
ax3d.set_zlabel('C')
plt.show()
```
